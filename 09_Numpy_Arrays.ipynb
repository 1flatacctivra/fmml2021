```python
# Array Attributes - ndim, size, shape, dtype, strides
```

```python
# ndim
w = np.array([[1],[2]])
x = np.array([1, 2])
y = np.array([[1.0, 2.0, 3.3], [4.2, 1.2, 1.1]]) 
z = np.arange(10)
print (w.ndim, x.ndim, y.ndim, z.ndim)
```

```python
print (z)
```

```python
# adding dimensions of an array
```

```python
new = np.array(y, ndmin=3)
```

```python
new
```

```python
# size
print (w.size, x.size, y.size, z.size)
```

```python
# shape
print (w.shape, x.shape, y.shape, z.shape)
```

```python
# dtype
x = np.array([1, 2])  # Let numpy choose the datatype
y = np.array([[1.0, 2.0, 3.3], [4.2, 1.2, 1.1]])  # Let numpy choose the datatype
z = np.arange(10, dtype=np.int16)  # Force a particular datatype

print(x.dtype, y.dtype, z.dtype)
```

```python
# strides
print (x.strides, y.strides, z.strides)
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,25,100)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x,y,'g+', label='sine')
plt.plot(x,z,'bo', label='cosine')
plt.xlabel('time')
plt.ylabel('range')
plt.title('Trigonometric')
plt.legend()
plt.show()
```
